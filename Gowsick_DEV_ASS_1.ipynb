# **1. Line Chart**

```python
import matplotlib.pyplot as plt
import pandas as pd

data = {'Date': pd.date_range(start='2023-01-01', periods=10, freq='D'),
        'Value': [10, 5, 20, 25, 30, 28, 32, 34, 40, 20]}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Value'], marker='o')
plt.title('Line Chart Example')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

Explanation : A line chart displays data points connected by straight lines. It is useful for showing trends over time or continuous data.

# **2. Bar Chart**

```python

import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [10, 35, 15, 30]

plt.figure(figsize=(10, 6))
plt.bar(categories, values, color='violet')
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

Explanation: A bar chart displays data using rectangular bars. It is effective for comparing different categories.

# **3. Scatter Plot**

```python
# Sample Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 6]

plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='red')
plt.title('Scatter Plot Example')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

Explanation: A scatter plot shows individual data points on a two-dimensional plane. It is useful for examining relationships between two variables.

# **4. Bubble Chart**

```python

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 8.5]
sizes = [100, 400, 600, 800, 1000]

plt.figure(figsize=(10, 6))
plt.scatter(x, y, s=sizes, alpha=0.5, color='blue')
plt.title('Bubble Chart Example')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

Explanation: A bubble chart displays three dimensions of data: two numerical variables as the X and Y axes and a third variable as the size of the bubbles

# **5. Scatter Plot Using Seaborn**

```python

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [2, 3, 5, 7, 11]})

plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='x', y='y', color='blue')
plt.title('Scatter Plot Using Seaborn')
plt.show()
```

Explanation: Seaborn is a Python library based on Matplotlib that provides a high-level interface for drawing attractive statistical graphics.

# **6. Area Plot and Stacked Plot**

```python

import matplotlib.pyplot as plt
import pandas as pd

data = pd.DataFrame({
    'Date': pd.date_range(start='2023-01-01', periods=10, freq='D'),
    'A': [10, 15, 20, 25, 30, 28, 32, 34, 40, 42],
    'B': [5, 7, 8, 6, 7, 5, 8, 9, 10, 11]
})

plt.figure(figsize=(10, 6))
plt.fill_between(data['Date'], data['A'], color="violet", alpha=0.4, label='A')
plt.fill_between(data['Date'], data['B'], color="green", alpha=0.4, label='B')
plt.title('Area Plot Example')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```

Explanation: An area plot displays the cumulative magnitude of a variable over time, while a stacked plot shows the breakdown of a variable across categories.

# **7. Pie Chart**

```python
import matplotlib.pyplot as plt

labels = ['Category A', 'Category B', 'Category C']
sizes = [40, 30, 450]

plt.figure(figsize=(10, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Aravi Pie Chart ')
plt.show()
```

Explanation: A pie chart displays data as slices of a circle, representing proportions of a whole.

# **8. Table Chart**

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.DataFrame({
    'S.no': [1, 2,3,4,5,6],
    'Team members Name': ['Aravind','Nandhavel','Edison','Kishore kumar','Harish venkat','Mukesh kanna']
})
fig, ax = plt.subplots(figsize=(6, 2))
ax.axis('tight')
ax.axis('off')
ax.table(cellText=data.values, colLabels=data.columns, cellLoc='center', loc='center')
plt.title('Team OG ')
plt.show()
```

Explanation: A table chart displays data in tabular form, useful for presenting raw data.

# **9. Polar Chart**

```python
import matplotlib.pyplot as plt
import numpy as np

labels=np.array(['A', 'B', 'C', 'D'])
values=np.array([7, 13, 20, 5])
angles=np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()

values=np.concatenate((values,[values[0]]))
angles+=angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, values, color='blue', alpha=0.5)
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)
plt.title('Polar Chart Example')
plt.show()
```

Explanation: A polar chart displays data on a circular grid, useful for showing data that is cyclical in nature.

# **10. Histogram**

```python

import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color='green', edgecolor='black')
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

Explanation: A histogram displays the distribution of numerical data by dividing it into bins and counting the number of observations in each bin.

# **11. Lollipop Chart**

```python

import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

plt.figure(figsize=(10, 6))
markerline, stemlines, baseline = plt.stem(categories, values)
baseline.set_visible(False)
plt.title('Lollipop Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

Explanation: A lollipop chart is similar to a bar chart but uses lines to connect the bar tops to a baseline, providing a cleaner look.
